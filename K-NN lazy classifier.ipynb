```python
import os
import numpy as ny
from numpy import *
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
addr = "C:\\Users\\BIT\\Data mining assignments\\iris.data.txt"

iris_data_raw = ny.genfromtxt(addr, delimiter = ",",usecols=(0,1,2,3))
iris_data_classes = ny.genfromtxt(addr,delimiter = ",",dtype = None,usecols=(4))

# Normalize data using Z-Norm
mean_of_data = ny.nanmean(iris_data_raw,axis = 0)
std_dev_of_data = ny.nanstd(iris_data_raw, axis = 0)
norm_iris_data = (iris_data_raw - mean_of_data)/std_dev_of_data

# Split the Data in test and train sets
test_lim = int(round(len(iris_data_raw)*.25))
train_lim = len(iris_data_raw) - test_lim
TEST_SET_DATA = norm_iris_data[:test_lim]
TEST_SET_CLASSES = iris_data_classes[:test_lim]
TRAIN_SET_DATA = norm_iris_data[test_lim:]
TRAIN_SET_CLASSES = iris_data_classes[test_lim:]

predictions = []
cand_1 = TEST_SET_DATA[0]

nearest_classes = []

dist_class_dict = {}



k = 3
for i in range(train_lim):
    x = ny.linalg.norm(cand_1-TRAIN_SET_DATA[i])
    dist_class_dict[x] = TRAIN_SET_CLASSES[i]

print ('Sort by keys:')
for key in sorted(dist_class_dict.keys()):
    nearest_classes.append(dist_class_dict[key])

k_nearest_classes = nearest_classes[:k]
counts = Counter(k_nearest_classes)
predictions.append(counts.most_common(1)[0][0])

predictions
```
